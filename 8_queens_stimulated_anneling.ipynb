```python
print("Akshara Singa\n1BM22CS029")
import random
import math

num_queens = 8

def count_conflicts(state):
    conflicts = 0
    n = len(state)
    for i in range(n):
        for j in range(i + 1, n):
            if state[i] == state[j] or abs(state[i] - state[j]) == j - i:
                conflicts += 1
    return conflicts

def initialize_board(size):
    return [random.randint(0, size - 1) for _ in range(size)]

def simulated_annealing(num_queens, max_steps=10000, initial_temp=100, cooling_factor=0.99):
    board_state = initialize_board(num_queens)
    current_conflicts = count_conflicts(board_state)
    temperature = initial_temp

    for step in range(max_steps):
        if current_conflicts == 0:
            print(f"Step {step}: Solution found!")
            print("Board Configuration:", board_state)
            return board_state

        row_to_modify = random.randint(0, num_queens - 1)
        new_column = random.randint(0, num_queens - 1)
        while new_column == board_state[row_to_modify]:
            new_column = random.randint(0, num_queens - 1)

        new_state = board_state[:]
        new_state[row_to_modify] = new_column
        new_conflicts = count_conflicts(new_state)
        conflict_delta = new_conflicts - current_conflicts

        if conflict_delta < 0 or math.exp(-conflict_delta / temperature) > random.random():
            board_state, current_conflicts = new_state, new_conflicts

        print(f"Step {step}: Temperature={temperature:.2f}, Current Conflicts={current_conflicts}, "
              f"Delta Conflicts={conflict_delta}, New Position for Row {row_to_modify} -> Column {new_column}")
        print("Board Configuration:", board_state)

        temperature *= cooling_factor

    print("No solution found within the maximum number of steps.")
    return None

final_solution = simulated_annealing(num_queens)
if final_solution:
    print("Final Solution:", final_solution)
else:
    print("No solution found within the maximum number of steps.")
```
