```python
print("Akshara Singa\n1BM22CS029")
# prompt: proof by resolution for
# Mother(Leela,Oshin)
# Alive(leela)
# !Mother(leela,Oshin) V Parent(leela,Oshin)
# !parent(leela,Oshin) V !alive(leela) V older(leela,oshin)
# to prove:older(leela,oshin)
# design code for the following

# Knowledge base
kb = [
    "Mother(Leela,Oshin)",
    "Alive(Leela)",
    "!Mother(Leela,Oshin) V Parent(Leela,Oshin)",
    "!Parent(Leela,Oshin) V !Alive(Leela) V Older(Leela,Oshin)"
]

# Goal to prove
goal = "Older(Leela,Oshin)"


def resolve(clause1, clause2):
    """
    Resolves two clauses.

    Returns:
      A new clause if resolution is possible, otherwise None.
    """
    literals1 = clause1.split(" V ")
    literals2 = clause2.split(" V ")

    for literal1 in literals1:
        for literal2 in literals2:
            if literal1[0] == "!" and literal1[1:] == literal2:
                # Resolve positive and negative literals
                new_clause = []
                for lit in literals1:
                    if lit != literal1:
                        new_clause.append(lit)
                for lit in literals2:
                    if lit != literal2:
                        new_clause.append(lit)
                return " V ".join(new_clause) if new_clause else ""  # Empty clause means contradiction

            elif literal1 == literal2[1:] and literal2[0] == "!":
                #Resolve positive and negative literals
                new_clause = []
                for lit in literals1:
                    if lit != literal1:
                        new_clause.append(lit)
                for lit in literals2:
                    if lit != literal2:
                        new_clause.append(lit)
                return " V ".join(new_clause) if new_clause else ""  # Empty clause means contradiction
    return None

def proof_by_resolution(kb, goal):
    """
    Performs proof by resolution.

    Args:
      kb: Knowledge base as a list of clauses.
      goal: Goal to prove.

    Returns:
      True if the goal is proven, False otherwise.
    """

    negated_goal = "!" + goal
    kb.append(negated_goal)  # Add the negation of the goal to the KB
    new_clauses = []

    while True:
      for i in range(len(kb)):
          for j in range(i + 1, len(kb)):
              resolvent = resolve(kb[i],kb[j])
              if resolvent == "":
                  print("Empty Clause generated. Proof Successful!")
                  return True

              if resolvent and resolvent not in kb and resolvent not in new_clauses:
                  print("New clause generated:", resolvent)
                  new_clauses.append(resolvent)

      if not new_clauses:
          print("No more new clauses. Proof Failed.")
          return False

      kb.extend(new_clauses) # add new clauses to kb
      new_clauses = [] # clear new clause list




if __name__ == "__main__":
    if proof_by_resolution(kb, goal):
        print(f"The goal '{goal}' is proven.")
    else:
        print(f"The goal '{goal}' could not be proven.")
```
